```python
import pandas as pd

# Cargar el archivo CSV
df_netflix_edaproducciones = pd.read_csv(r"C:\Users\fatim\Downloads\laboratorio-Pandas_Netflixproducciones\data.netflix.csv") 

# Configurar Pandas para mostrar todas las filas y columnas de nuestro conjunto de datos
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)

# Mostrar los DataFrame completos a través de display(df_netflix_edaproducciones) para darle un formato más visual que el que devuelve print
print("#### Dataframe de todas las producciones de Netflix ####")
display(df_netflix_edaproducciones_info = {
    "Head": df_netflix_edaproducciones.head(),  # First few rows of the dataset
    "Shape": df_netflix_edaproducciones.shape,  # Dimensions of the dataset
    "Columns": df_netflix_edaproducciones.columns.tolist(),  # List of column names
    "Summary": df_netflix_edaproducciones.describe(include='all')  # Resumen estadístico de nuestras variables numéricas y categóricas
})

display(df_netflix_edaproducciones_info)


```

```python

# Resumen general
print("Resumen de columnas:")
display(df_netflix_edaproducciones.info())

# Valores únicos en las principales columnas
print("\nValores únicos en 'type':")
display(df_netflix_edaproducciones['type'].value_counts())

print("\nDistribución por país:")
display(df_netflix_edaproducciones['country'].value_counts().head(10))

# Valores nulos
print("\nValores faltantes por columna:")
display(df_netflix_edaproducciones.isnull().sum())

# Resumen estadístico 
print("\nResumen estadístico de IMDb Score:")
display(df_netflix_edaproducciones['IMDB Score'].describe())

# Limpieza básica de datos
df_netflix_edaproducciones['country'] = df_netflix_edaproducciones['country'].replace('Unknown', pd.NA)

# Filtrar películas con alta calificación
high_rated = df_netflix_edaproducciones[df_netflix_edaproducciones['IMDB Score'] > 8]
print("\nPelículas con calificación IMDb mayor a 8:")
display(high_rated[['title', 'IMDB Score', 'country']])

```

```python
print("Todas las estadísticas descriptivas de las producciones de Netflix")
display(df_netflix_edaproducciones.describe(include='all').transpose()) #todas las columnas

display(df_netflix_edaproducciones.describe().T) #solo categóricas
```

```python
print(df_netflix_edaproducciones.columns.tolist())  # ¿cuáles son nuestras columnas?

print("Análisis detallado de las columnas clave para la estrategia de contenidos")

display(df_netflix_edaproducciones["type"].unique())
display(df_netflix_edaproducciones["country"].unique())
```

```python
producciones_mejorvaloradas = df_netflix_edaproducciones[df_netflix_edaproducciones['IMDB Score'] > 8.5]
contenidos_peorvalorados = df_netflix_edaproducciones[df_netflix_edaproducciones['IMDB Score'] < 5]

print("¿Qué contenidos han alcanzado mayor éxito?")

print("Producciones mejor valoradas:")
display(producciones_mejorvaloradas)

print("\nLas 5 producciones peor valoradas:")
display((contenidos_peorvalorados.head(5)))

print("¿Cuánto influye la duración en la calificación obtenida?")
runtime_vs_score = df_netflix_edaproducciones[['Runtime', 'IMDB Score']].corr()
print(runtime_vs_score)

# La columna rating contiene las clasificaciones de audiencia asignadas a cada contenido, las cuales indican para qué público es adecuado

print("¿Cuántas producciones hay para cada clasificación de audiencia?")
ratings_distribution = df_netflix_edaproducciones['rating'].value_counts()
print("\nDistribución de Clasificaciones por Audiencia:")
print(ratings_distribution)

print("¿Y qué clasificaciones tienden a tener mejores calificaciones?")
imdb_by_rating = df_netflix_edaproducciones.groupby('rating')['IMDB Score'].mean().sort_values(ascending=False)
print("\nCalificación Promedio por Clasificación:")
print(imdb_by_rating)

```

```python
print("Análisis en clave estratégica de internacionalización")
print("Idiomas más representadados")
top_languages = df_netflix_edaproducciones['Language'].value_counts().head(10)
print("\nIdiomas Más Representados:")
print(top_languages)

print("Éxito de calificaciones según idioma")
imdb_by_language = df_netflix_edaproducciones.groupby('Language')['IMDB Score'].mean().sort_values(ascending=False).head(10)
print("\nIdiomas con Mejores Calificaciones Promedio:")
print(imdb_by_language)
```

# Clasificaciones más frecuentes 

TV-MA con 159 contenidos dirigido a audiencias adultas. Es la clasificación predominante en el catálogo, lo que indica que Netflix está enfocado en producciones para adultos. Esta cifra refleja el segmento estratégico para el mercado de contenidos de Netflix.
TV-14 con 108 contenidos: apto para mayores de 14 años, es la segunda clasificación más común, centrada en el target de adolescentes.
PG-13 con 58 contenidos: está orientado también a un público joven-adulto.
TV-G (11 contenidos) y G (1 contenido). Se trata de contenido apto para todas las edades. Estas categorías están subrepresentadas, lo que podría indicar una baja prioridad en el contenido infantil o familiar.

```python
print("Años estrella por cantidadad de lanzamientos")
print("Número de contenido de estreno por año")
content_by_year = df_netflix_edaproducciones['release_year'].value_counts().sort_index()
print("\nContenido Agregado por Año:")
print(content_by_year)

print("Calificación promedio según su fecha de estreno en la plataforma")
avg_imdb_by_year = df_netflix_edaproducciones.groupby('release_year')['IMDB Score'].mean().sort_index()
print(avg_imdb_by_year)
```
